```python
Q1 = df["Boxes Shipped"].quantile(0.25)
Q3 = df["Boxes Shipped"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df["Boxes Shipped"] < lower_bound ) & (df ["Boxes Shipped"] > upper_bound)]
print(outliers.shape[0])
```

```python
Q1 = df["Amount ($)"].quantile(0.25)
Q3 = df["Amount ($)"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df["Amount ($)"] < lower_bound ) & (df ["Amount ($)"] > upper_bound)]
print(outliers.shape[0])
```

```python
#ANALYSE
```

```python
df.sample(10)
```

```python
df["Month_Year"] = df["Date"].dt.to_period("M")
```

```python
monthly_sales = df.groupby("Month_Year")["Amount ($)"].sum().reset_index()
monthly_sales["Month_Year"] = monthly_sales["Month_Year"].astype(str)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x="Month_Year", y="Amount ($)", marker="o")
plt.xlabel("Month")
plt.ylabel("Total sales ($)")
plt.title("Monthly sales trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
country_sales = df.groupby("Country")["Amount ($)"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=country_sales.head(5).values, y=country_sales.head(5).index, palette="viridis")
for i, value in enumerate(country_sales.head(5).values):
    plt.text(value + 200, i, f"${value:,.2f}",  va="center")
plt.xlabel("Total sales ($)")
plt.ylabel("Country")
plt.title("Top Countries by Total Sales")
max_value = country_sales.head(1).values[0]
plt.xticks(np.arange(0, max_value + 4000, 4000))
plt.tight_layout()
plt.show()
```

```python
sales_by_person = df.groupby("Sales Person")["Amount ($)"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=sales_by_person.head(5).values, y=sales_by_person.head(5).index, palette="Blues_d")
for i, value in enumerate(sales_by_person.head(5).values):
    plt.text(value + 10, i, f"${value:,.2f}",  va="center")
plt.xlabel("Total sales ($)")
plt.ylabel("Sales Persons")
plt.title("Top Sales Persons by Total Sales")
max_value = sales_by_person.head(1).values[0]
plt.xticks(np.arange(0, max_value + 4000, 4000))
plt.tight_layout()
plt.show()
```

```python
product_sales_volume = df.groupby("Product")["Boxes Shipped"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales_volume.head(7).values, y=product_sales_volume.head(7).index, palette="Greens_d")
for i, value in enumerate(product_sales_volume.head(7).values):
    plt.text(value + 1, i, f"{value}",  va="center")
plt.xlabel("Total Boxes Shipped")
plt.ylabel("Product")
plt.title("Top Products by Quantity sold")
plt.tight_layout()
plt.show()
```

```python
top_products= product_sales_volume.head(7)
plt.figure(figsize=(8, 8))
plt.pie(top_products,
        labels=top_products.index,
        autopct='%1.1f%%',
        startangle=140,
colors=sns.color_palette("pastel"))
plt.title("Top Products by Quantity Sold")
plt.axis("equal")
plt.show()
```

```python
product_sales_amount = df.groupby("Product")["Amount ($)"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales_amount.head(7).values, y=product_sales_amount.head(7).index, palette="YlGnBu")
for i, value in enumerate(product_sales_amount.head(7).values):
    plt.text(value + 10, i, f"${value:,.2f}",  va="center")
plt.xlabel("Total Sales")
plt.ylabel("Product")
plt.title("Top Products by Total Sales")
max_value = product_sales_amount.head(1).values[0]
plt.xticks(np.arange(0, max_value + 4000, 4000))
plt.tight_layout()
plt.show()
```
